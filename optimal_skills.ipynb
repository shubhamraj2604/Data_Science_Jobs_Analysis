### WHAT IS THE MOST OPTIMAL SKILL TO LEARN FOR DATA ANALYST?

- Grouping skills to determine median salary and likelihood of being in posting

```python
# Importing Libraries
import ast
import pandas as pd
import seaborn as sns
from datasets import load_dataset
import matplotlib.pyplot as plt  
import seaborn as sns

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df
```

```python
df_DA_US = df[(df['job_country'] == 'United States') & (df['job_title_short'] == 'Data Analyst')].copy()
df_DA_US
```

```python
df_DA_US_exploaded = df_DA_US.explode('job_skills')
```

```python
df_DA_skills = df_DA_US_exploaded.groupby('job_skills')['salary_year_avg'].agg(['count' , 'median']).sort_values(by='count' , ascending=False)
df_DA_skills = df_DA_skills.rename(columns={'count':'skill_count' , 'median' : 'median_salary'})
df_DA_skills

```

```python
DA_job_count = len(df_DA_US)
df_DA_skills['skill_percent'] = df_DA_skills['skill_count'] / DA_job_count * 100
df_DA_skills
```
